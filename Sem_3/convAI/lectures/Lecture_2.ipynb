Text Generation using GPT 2

```python
# Install the transformers library
!pip install transformers

# Import necessary libraries
from transformers import GPT2LMHeadModel, GPT2Tokenizer

# Load the pre-trained GPT-2 model and tokenizer
model_name = "gpt2"  # You can use "gpt2-medium", "gpt2-large", "gpt2-xl" for larger models
tokenizer = GPT2Tokenizer.from_pretrained(model_name)
model = GPT2LMHeadModel.from_pretrained(model_name)

# Function to generate text
def generate_text(prompt, max_length=100):
    # Encode the input text
    inputs = tokenizer.encode(prompt, return_tensors="pt")
    # Generate text
    outputs = model.generate(inputs, max_length=max_length, num_return_sequences=1, no_repeat_ngram_size=2, early_stopping=True)
    # Decode the generated text
    text = tokenizer.decode(outputs[0], skip_special_tokens=True)
    return text

# Input prompt
prompt = "Once upon a time"

# Generate and display the text
generated_text = generate_text(prompt)
print(generated_text)
```

  State-based Dialogue System

```python
# Define states
INITIAL_STATE = "initial"
ASK_INFO_STATE = "ask_info"
PROVIDE_INFO_STATE = "provide_info"
END_STATE = "end"

# Define a dictionary to hold state transitions
state_transitions = {
    INITIAL_STATE: ASK_INFO_STATE,
    ASK_INFO_STATE: PROVIDE_INFO_STATE,
    PROVIDE_INFO_STATE: END_STATE
}

# Define a dictionary to hold state actions
state_actions = {
    INITIAL_STATE: "Hello! How can I assist you today?",
    ASK_INFO_STATE: "Can you please provide more details?",
    PROVIDE_INFO_STATE: "Thank you for the information. Here is what I found: [info].",
    END_STATE: "Goodbye! Have a great day!"
}

# Define a function to simulate the dialogue system
def dialogue_system():
    current_state = INITIAL_STATE
    while current_state != END_STATE:
        print(state_actions[current_state])
        user_input = input()
        current_state = state_transitions.get(current_state, END_STATE)
    print(state_actions[END_STATE])

# Run the dialogue system
dialogue_system()
```

Rule Based System

```python
# Define rules
rules = {
    "hello": "Hi, how can I help you today?",
    "account balance": "Please provide your account number.",
    "goodbye": "Goodbye! Have a great day!"
}

# Define a function to simulate the rule-based dialogue system
def rule_based_system():
    while True:
        user_input = input().lower()
        response = rules.get(user_input, "I'm sorry, I didn't understand that.")
        print(response)
        if user_input == "goodbye":
            break

# Run the rule-based dialogue system
rule_based_system()
```

```python
import spacy
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import make_pipeline

# Sample training data
training_data = [
    ("I want to book a flight to New York on June 5th", "book_flight"),
    ("Book a flight to London for tomorrow", "book_flight"),
    ("Can I get a flight to Paris on Monday", "book_flight"),
    ("What's the weather in San Francisco", "check_weather"),
    ("How's the weather in New York today", "check_weather"),
    ("Tell me the weather in Los Angeles", "check_weather")
]

# Split the data into inputs and labels
texts, labels = zip(*training_data)

# Create a text classification pipeline
vectorizer = CountVectorizer()
classifier = LogisticRegression()
model = make_pipeline(vectorizer, classifier)

# Train the model
model.fit(texts, labels)

# Test the model
test_text = "I need to book a flight to Tokyo next week"
predicted_intent = model.predict([test_text])[0]
print(f"Predicted Intent: {predicted_intent}")
```

```python
import spacy

# Load spaCy model
nlp = spacy.load("en_core_web_sm")

# Define a function for slot filling
def extract_slots(text, intent):
    doc = nlp(text)
    slots = {}
    if intent == "book_flight":
        for ent in doc.ents:
            if ent.label_ == "GPE":
                slots["location"] = ent.text
            elif ent.label_ == "DATE":
                slots["date"] = ent.text
    elif intent == "check_weather":
        for ent in doc.ents:
            if ent.label_ == "GPE":
                slots["location"] = ent.text
    return slots

# Test slot filling
slots = extract_slots(test_text, predicted_intent)
print(f"Extracted Slots: {slots}")
```
